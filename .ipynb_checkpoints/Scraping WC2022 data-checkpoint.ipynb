```python
import pandas as pd
import numpy as np
import time
```

```python
page_url = 'https://fbref.com/en/comps/1/World-Cup-Stats'
```

```python
league_table = pd.read_html(page_url, match='League Table')
league_table[0]
```

```python
import requests
from bs4 import BeautifulSoup
import re
request = requests.get(page_url)
soup = BeautifulSoup(request.text)
```

```python
teams = [x['href'] for x in soup.select('#results202210_overall td.left a')]
teams
```

```python
roster_table = pd.read_html('https://fbref.com' + teams[0], match='Roster')
time.sleep(3)
standard_stats_table = pd.read_html('https://fbref.com' + teams[0], match='Standard Stats')
time.sleep(3)
shooting_table = pd.read_html('https://fbref.com' + teams[0], match='Shooting')
time.sleep(3)
passing_table = pd.read_html('https://fbref.com' + teams[0], match='Passing')
time.sleep(3)
miscellaneous_stats_table = pd.read_html('https://fbref.com' + teams[0], match='Miscellaneous Stats')
```
